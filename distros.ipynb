```python
from scipy.stats import norm
from numpy import sqrt

n = 60
p = 99/100

mu = n*p
sig = sqrt(n*p*(1-p))

a = 0 - 1/2
b = 60 - 1/2

res = 1-norm.cdf(b, loc=mu, scale=sig)

res
```

```python
# OBSERVACIN: La suma de exponenciales independientes no retorna una exponencial

nums1 = expon.rvs(scale=2, size=1000)

plt.hist(nums1, bins=100)
plt.show()

nums2 = expon.rvs(scale=2, size=1000)
plt.hist(nums2, bins=100)
plt.show()

nums3 = nums1 + nums2
plt.hist(nums3, bins=100)
plt.show()

nums4 = expon.rvs(scale=4, size=1000)
plt.hist(nums4, bins=100)
plt.show()

print()
```
